```python
import datetime as dt
import pandas as pd
import alpaca_trade_api as tradeapi
import time

API_KEY='xxxxxxxxxxxx'
SECRET_KEY='xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx'
BASE_URL='https://paper-api.alpaca.markets'

def calculate_bollinger_bands(prices, window=20, num_std_devs=2):
    sma = prices['close'].rolling(window=window).mean()
    std_dev = prices['close'].rolling(window=window).std()
    upper_band = sma + num_std_devs * std_dev
    lower_band = sma - num_std_devs * std_dev

    return sma, upper_band, lower_band

def execute_order(symbol, qty, side, order_type='market', time_in_force='gtc'):
    try:
        order = rest_api.submit_order(
            symbol=symbol,
            qty=qty,
            side=side,
            type=order_type,
            time_in_force=time_in_force
        )
        print(f"Order submitted: {order}")
    except Exception as e:
        print(f"Error: {e}")

def bollinger_bot(symbol, qty):
    start_date = (dt.datetime.now() - dt.timedelta(days=365)).isoformat() + 'Z'
    end_date = (dt.datetime.now() - dt.timedelta(minutes=15)).isoformat() + 'Z'
    df = pd.DataFrame(rest_api.get_bars(symbol, timeframe, start=start_date, end=end_date).df)
    df = df.iloc[::-1]

    last_close = df.iloc[-1]['close']
    sma, upper_band, lower_band = calculate_bollinger_bands(df)
    
    if last_close < lower_band.iloc[-1]:
        print("Buy signal")
        execute_order(symbol, qty, 'buy')
    elif last_close > upper_band.iloc[-1]:
        print("Sell signal")
        execute_order(symbol, qty, 'sell')
    else:
        print("No signal")

rest_api = tradeapi.REST(API_KEY, SECRET_KEY, BASE_URL, api_version='v2')
timeframe = '1D'

stock_list = [
    'AAPL', 'MSFT', 'AMZN', 'GOOGL', 'TSLA',
    'META', 'NVDA', 'NFLX', 'ADBE', 'CRM',
    'V', 'MA', 'PYPL', 'JPM', 'GS',
    'UNH', 'PFE', 'JNJ', 'PG', 'KO'
]

for stock in stock_list:
    symbol = stock
    qty = 1
    bollinger_bot(symbol, qty)
    time.sleep(1)
```
